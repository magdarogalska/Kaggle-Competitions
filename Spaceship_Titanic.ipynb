```python
cols = list(train_cat.columns)

for j in cols:
    train_cat[j] = train_cat[j].fillna('nan')
    

    for row in range(train_cat.shape[0]):
        if train_cat.iloc[row][j]=='nan':
            group=train_cat.iloc[row]['Group']
            groups = train_cat[j][train_cat['Group']==group]
            for i in groups:
                if(i!='nan'):
                    train_cat.at[row,j]= i
                    break
            if train_cat.iloc[row][j]=='nan':
                train_cat.at[row,j]=None
                
for j in cols:
    test_cat[j] = test_cat[j].fillna('nan')
    

    for row in range(test_cat.shape[0]):
        if test_cat.iloc[row][j]=='nan':
            group=test_cat.iloc[row]['Group']
            groups = test_cat[j][test_cat['Group']==group]
            for i in groups:
                if(i!='nan'):
                    test_cat.at[row,j]= i
                    break
            if test_cat.iloc[row][j]=='nan':
                test_cat.at[row,j]=None
```

```python
X = pd.read_csv('train_transformed6.csv')
y=pd.read_csv('y.csv')
test = pd.read_csv('test_transformed6.csv')
```

```python
X = X.drop(['PassengerId','Group'], axis=1)
test = test.drop(['PassengerId','Group'], axis=1)
```

```python
X
```

```python
from sklearn.ensemble import RandomForestClassifier
y = y.replace({False:0, True:1})
```

```python
from sklearn.model_selection import train_test_split

train_X, test_X, train_y, test_y = train_test_split(X,y, random_state=42, test_size=0.2)
rfc = RandomForestClassifier(random_state=42)
rfc.fit(train_X, train_y)
rfc_pred = rfc.predict(test_X)
```

```python
from sklearn.metrics import f1_score, accuracy_score
rfc_acc = accuracy_score(test_y, rfc_pred)
rfc_f1 = f1_score(test_y, rfc_pred)
print('Accuracy : {:.3f} \nF1 : {:.3f}'.format(rfc_acc, rfc_f1))
```

```python
rfc_acc
```

```python
from sklearn.neighbors import KNeighborsClassifier
neighbors = [2,4,6,8,10,12,16,20]
for i in neighbors:
    knn = KNeighborsClassifier(n_neighbors= i)
    knn.fit(train_X, train_y)
    knn_pred = knn.predict(test_X)
    knn_acc = accuracy_score(test_y, knn_pred)
    rfc_f1 = f1_score(test_y, knn_pred)
    print("Neighbors: {}\nAccuracy : {:.3f} \nF1 : {:.3f}".format(i, rfc_acc, rfc_f1))
```

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(random_state=42, max_iter=1000, warm_start=True)
mlp.fit(train_X, train_y)
mlp_pred = mlp.predict(test_X)
mlp_acc = accuracy_score(test_y, mlp_pred)
mlp_f1 = f1_score(test_y, mlp_pred)
print('Accuracy : {:.3f} \nF1 : {:.3f}'.format(mlp_acc, mlp_f1))
```

```python
from sklearn.decomposition import PCA

comp = range(2,15)
best_acc = 0
best_f1 =0
for i in comp:
    
    pca = PCA(n_components=i)
    X_pca = pca.fit_transform(X)
    X_pca= pd.DataFrame(X_pca, index=X.index)
    train_X, test_X, train_y, test_y = train_test_split(X_pca,y, random_state=42, test_size=0.2)
    rfc = RandomForestClassifier(random_state=42)
    rfc.fit(train_X, train_y)
    rfc_pred = rfc.predict(test_X)
    rfc_acc = accuracy_score(test_y, rfc_pred)
    rfc_f1 = f1_score(test_y, rfc_pred)
    if rfc_acc>best_acc:
        best_acc=rfc_acc
    if rfc_f1>best_f1:
        best_f1=rfc_f1
    print('\nNum of components: {}\nAccuracy : {:.3f} \nF1 : {:.3f}\n\n\n'.format(i,rfc_acc, rfc_f1))
    
    
```
